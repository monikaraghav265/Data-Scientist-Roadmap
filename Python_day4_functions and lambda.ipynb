```python
#lambda function 
sq=lambda x:x**2
print(sq(2))
```

```python
sq=lambda x,y:x+y
print(sq(2,3))
```

```python
#Check even/odd
def is_even(num):
    if num%2==0:
        print("yes,even")
    else:
        print("No") 
is_even(3) 
    
```

```python
num=int(input("Enter any number"))     #Factorial of a number 
prod=1
for i in range (1,num+1):
    prod=prod*i
print(prod) 
```

```python
#Factorial of a number by user define function 
def fact(num):
    prod=1
    for i in range(1,num+1):
        prod=prod*i
    return(prod) 
fact(5)
```

```python
#Sum of number in a list 
def sum_list(lst):
    sum=0
    for i in lst:
        sum=sum+i
    return(sum) 
sum_list([2,3,4]) 
```

```python
#Lambda function (Double number) 
num=lambda x:x*2
print(num(2))
```

```python
#Double number in a list   (using map func. and lambda func.) 
lst=[2,3,4,5]
doubled=list(map(lambda x:x*2,lst))     #map function always have 2 arguments 
print(doubled) 
```
