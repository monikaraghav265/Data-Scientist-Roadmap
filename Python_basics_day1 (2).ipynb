```python
#Simple Calculator(+,-,*,/) 
f1=int(input("Enter first number"))
f2=int(input("Enter second number")) 
op=input("Enter operator:(+,-,*,/)") 
if op=="+":
    print("Result",f1+f2) 
elif op=="-":
    print("Result",f1-f2) 
elif op=="*":
    print("Result",f1*f2) 
elif op=="/":
    print("Result",f1/f2)
else:
    print("Invalid operator")   
```

```python
#Reverse a number  (%=last digit nikalta h, //=last digit htata h , *10=digit ko left shift krta h 
rev=0
num=123
while num>0:
    digit=num%10
    rev=rev*10+digit
    num=num//10
print(rev) 
```
